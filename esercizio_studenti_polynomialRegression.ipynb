```python
# Import the required libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Read the CSV file:
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/AI/student-mat.csv", sep=';')
data
```

```python
data['sex'] = data['sex'].replace({'M':0, 'F':1})
data
```

```python
# Consider features we want to work on:
X = data.drop(['G3','school','Pstatus','address','famsize', 'Mjob', 'Fjob','reason', 'guardian' ], axis = 'columns').replace({'no':0, 'yes':1}).to_numpy()
X
```

```python
Y = data['G3'].to_numpy()
Y
```

```python
from sklearn.model_selection import train_test_split # suddivide le osservazioni nei due insiemi

X_train, X_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size = 0.33,
                                                    random_state = 123) # fix random seed for replicability
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
print(X_train[:10])
print(f'y_train = {y_train[:10]}')
```

```python
poly = PolynomialFeatures(degree = 3)
X_train_poly = poly.fit_transform(X_train)
```

```python
modello = LinearRegression()
modello.fit(X_train_poly, y_train)
```

```python
X_test_poly = poly.transform(X_test)
predict = modello.predict(X_test_poly)
```

```python
print(f'dati predetti dal metodo = {predict[:10].round(decimals=2)}') # dati predetti dal metodo
print(f'dati reali presenti nel dataset di test = {y_test[:10]}') # dati effettivi presenti nel test set
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# The coefficients
print('Coefficients: \n', modello.coef_)

# The mean squared error
print('Mean squared error: %.2f' %mean_squared_error(y_test, predict))

# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f' %r2_score(y_test, predict))
```

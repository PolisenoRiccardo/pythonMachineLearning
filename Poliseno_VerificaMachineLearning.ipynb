```python
xH = hardware.drop([0, 1,9], axis = 'columns').to_numpy()
yH = hardware[9].to_numpy()
print(xH, yH)
```

```python
bankdrop = bank.drop(bank[(bank['pdays'] == -1)].index)
xB = bankdrop.drop(['y', 'job', 'education', 'contact',  'poutcome', 'marital', 'month'], axis = 'columns').replace({'yes': 1, 'no': 0}).to_numpy()
yB = bankdrop['y'].to_numpy()
print(xB, yB)
```

4. un modello di classificazione o regressione (a seconda di ciÃ² che Ã¨ richiesto)

```python
from sklearn.model_selection import train_test_split

XH_train, XH_test, yH_train, yH_test = train_test_split(xH, yH,
                                                    test_size = 0.33,
                                                    random_state = 123)

XB_train, XB_test, yB_train, yB_test = train_test_split(xB, yB,
                                                    test_size = 0.33,
                                                    random_state = 123)
```

```python
XH_train.shape,yH_train.shape,XH_test.shape,yH_test.shape
```

```python
XB_train.shape,yB_train.shape,XB_test.shape,yB_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(XB_train, yB_train)
predictB = neigh.predict(XB_test)
print(f'dati predetti dal metodo =                {predictB[:10]}')
print(f'dati reali presenti nel dataset di test = {yB_test[:10]}')
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()

regr.fit(XH_train,yH_train)
predictHL = regr.predict(XH_test)

print(f'dati predetti dal metodo = {predictHL[:10].round(decimals=2)}')
print(f'dati reali presenti nel dataset di test = {yH_test[:10]}')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly2 = PolynomialFeatures(degree = 2)
XH_train_poly2 = poly2.fit_transform(XH_train)

modello = LinearRegression()
modello.fit(XH_train_poly2, yH_train)

XH_test_poly2 = poly2.transform(XH_test)
predictHP2 = modello.predict(XH_test_poly2)
```

```python
poly3 = PolynomialFeatures(degree = 3)
XH_train_poly3 = poly3.fit_transform(XH_train)

modello = LinearRegression()
modello.fit(XH_train_poly3, yH_train)

XH_test_poly3 = poly3.transform(XH_test)
predictHP3 = modello.predict(XH_test_poly3)
```

5. una valutazione del modello con una breve descrizione della sua bontÃ  nell'effettuare previsioni

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(yB_test, predictB))

'''
Il modello funziona abbastanza bene con i risultati negativi, ma con quelli positivi ne sbaglia uno su due, quindi non Ã¨ buono, Ã¨ come se indovinasse casualmente.
'''
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# The coefficients
print('Coefficients: \n', regr.coef_)

# The mean squared error
print('Mean squared error: %.2f' %mean_squared_error(yH_test, predictHL))

# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f' %r2_score(yH_test, predictHL))
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)

# The mean squared error
print('Mean squared error: %.2f' %mean_squared_error(yH_test, predictHP2))

# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f' %r2_score(yH_test, predictHP2))

# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f' %r2_score(yH_test, predictHP3))
```
